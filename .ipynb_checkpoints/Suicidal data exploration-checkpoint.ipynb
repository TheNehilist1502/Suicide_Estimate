##### Economical terms:
######                                   
###### *HDI*(Human development index): it is a statistic composite index of life expectancy, education (Literacy Rate, Gross Enrollment Ratio at different levels and Net Attendance Ratio), and per capita income indicators, which are used to rank countries into four tiers of human development. A country scores a higher HDI when the lifespan(The expected time to live) is higher, the education level is higher, and the gross national income GNI (PPP) per capita (the rate of individual income) is higher.
###### *gdp*=*ppp*: (gross domestic product): the purchasing power parity (PPP which is the affordability to buy final goods and needs) value of all final goods and services produced within a country in a given year. When it is divided by the average population for the same year. Then it is *gdp per capita *
######                                  
##### Generation distribution terms:
######                                                
###### GI generation: Born(1900-1927)
###### Silent: Born(1928-1945)
###### Boomers: Born(1946-1964)
###### Generation x: Born (1965-1980)

#### 1. Feature Engineering
In this process, I will get the best of data by getting new features out of old ones if possible or if needed.

```python
df.head()
```

Thinking about the feature that might be added as a domain-knowledge one. adding latitude and longitude would help as in countries that is not listed in our dataset would be handled better and expected with a better proficieny as near countries shall have som sort of similar circumstances that may affect civils and their suicidal rates.

```python
from  geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='Essam')
```

```python
#Getting two lists for lat and lang
lat=[]
lang=[]
counter=0
for i in df['country'].value_counts():
    country =df['country'].value_counts().index[counter]
    loc = geolocator.geocode(country)
    lat.append(loc.latitude)
    lang.append(loc.longitude)
    counter+=1
```

The idea to be done is to get a dictionary with three values where column names(Country,lan,lang) are the keys & the lists that lists (value_counts of *countries, lan, lang*) are the values then turning it onto a dataframe. Then, this data frame (let us call it lat_lang) would be merged to our main df on the column *country* which is common between two of them.

```python
data={'country':df['country'].value_counts().index, 'lat':lat, 'lang': lang}
```

```python
lat_lang=pd.DataFrame(data)
```

```python
lat_lang
```

```python
df=pd.merge(df,lat_lang,how='outer',on='country')
```

```python
df.head()
```

```python
df.info()
```
